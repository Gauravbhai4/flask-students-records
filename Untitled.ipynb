```python
import sqlite3 as sql
def sql_qu(query):
    dbname='record.db'
    con=sql.connect(dbname)
    cursor=con.cursor()
    cursor.execute(query)
    result=cursor.fetchall()
    con.commit()
    cursor.close()
    con.close()
    return result
result=sql_qu("SELECT * FROM student_marks")
print(result)
```
